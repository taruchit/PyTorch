Example of small neural network

```python
import torch.nn as nn
```

```python
x=torch.tensor(6.7)
y=torch.tensor(0.0)
```

```python
x
```

```python
y
```

```python
w=torch.tensor(1.0, requires_grad=True)
b=torch.tensor(0.0, requires_grad=True)
```

```python
w
```

```python
b
```

```python
#Forward pass
z=w**x+b
```

```python
#Sigmoid
y_pred=torch.sigmoid(z)
```

```python
#Binary cross entropy loss
criterion=nn.BCELoss()
loss=criterion(y_pred, y)
```

```python
loss
```

```python
loss.backward()
```

```python
w.grad
```

```python
b.grad
```

```python
loss
```

```python
#Clearing gradients
w.grad.zero_()
b.grad.zero_()
```
